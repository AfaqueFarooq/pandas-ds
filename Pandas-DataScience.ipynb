# Filtering Data


We take pokemons only that contains 'pi' string in it

```python
#lets import regular expressions
#flags=re.I is for ignore case(igonres if alphabets are capital or small)
import re

df[df['Name'].str.contains('pi[a-z]*',flags=re.I)]
```

```python
#with multiple conditions

df = df[(df['Type 1'].isin(['Fire','Water','Electric'])) & (df['Total']>320) & (df['Attack']>54)].head(50)
```

```python
print(type(df))  # Should print <class 'pandas.core.frame.DataFrame'>
```

```python
#Conditional Changes
#Case: when Type 1= Fire then Type 2 = Dragon 
#Case: when Type1= Water then Type 2 = Ice
#Case:when Type1=Electric then Type2=Thunder

df.loc[df['Type 1']=='Fire','Type 2']='Dragon'

df.loc[df['Type 1']=='Water','Type 2']='Ice'
df.loc[df['Type 1']=='Electric','Type 2']='Thinder'
df
```

```python
df['UpdatedName'] = df['Name'].apply(lambda x: 'Charizard' if 'Charizard' in x else 
                                     ('Blastoise' if 'Blastoise' in x 
                                      else ('Slowbro' if 'Slowbro' in x else ('Gyarados' if 'Gyarados' in x else 
                                                                             ('Ampharos' if 'Ampharos' in x else x)))))

df
```

```python
#to create new column
#lets say we want both the types in one column

df['Types'] = df['Type 1'] + ','+df['Type 2']

df.head()
```

```python
#concatenating Type 1 and Type 2 with a comma, but if Type 2 is NaN, the resulting Types value will also include NaN. To handle this, you can use the fillna('') method to replace NaN with an empty string before concatenation, so it doesn't appear in the result.

#df['Types'] = df['Type 1'] + ','+df['Type 2'].fillna('')

#df.head()
```

```python
#now lets drop Type 1 and Type 2 as we dont need it anymore

df = df.drop(columns=[('Type 1'),('Type 2')])
```

```python
df
```

```python
#now lets try without dropping colmumn instead we take only columns we need

df = df[['UpdatedName','Total','HP','Generation','Types']]

df.head()
```

```python
#renaming UpdateName Column

df = df.rename(columns={'UpdatedName':'Name'})

df
```

```python
 #lets save the new changes in csv
 #we can save output as excel using to_excel()

df = df.to_csv('pokemon_modified.csv')
```
